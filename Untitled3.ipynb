## Exploratory Data Analysis

```python
bike_df.describe(include = "all")
```

```python
year_dict = {0: "2011", 1 :"2012"}
bike_df["yr"] = bike_df["yr"].map(year_dict)
bike_df
```

```python
#Mengubah data kategori workingday dalam bentuk string
workingday_dict = {0: "workingday", 1: "weekend/holiday"}
bike_df["workingday"] = bike_df["workingday"].map(workingday_dict)
bike_df
```

```python
# Mengubah data kategori cuaca dalam bentuk string
weather_dict = {1: "clear", 2: "cloudy", 3: "light snow/rain", 4: "snow/heavy rain"}
bike_df["weathersit"] = bike_df["weathersit"].map(weather_dict)
bike_df
```

```python
# Kelompok peminjaman sepeda berdasarkan cuaca
bike_df.groupby(by = "weathersit").agg({
    "cnt" : "sum",
    "casual" : "sum",
    "registered" : "sum"
})
```

```python
season_dict = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
bike_df['season'] = bike_df['season'].map(season_dict)
bike_df
```

```python
# Kelompok peminjaman sepeda berdasarkan musim
bike_df.groupby(by = "season").agg({
    "cnt" : "sum",
    "casual" : "sum",
    "registered" : "sum"
})
```

```python
# Kelompok peminjaman sepeda berdasarkan cuaca dan musim
bike_df.groupby(by = ["season", "weathersit"]).agg({
    "cnt" : "sum",
    "casual" : "sum",
    "registered" : "sum"
})
```

```python
# Kelompok peminjaman sepeda berdasarkan workingday dan musim
bike_df.groupby(by = ["workingday", "season"]).agg({
    "cnt" : "sum",
    "casual" : "sum",
    "registered" : "sum"
})
```

## Visual & Explanarory Analysis

* 1. Bagaimana frekuensi peminjaman sepeda untuk setiap bulan pada tahun 2011 dan 2012?

```python
bike_df_year = bike_df.groupby(by = ["yr","mnth"]).agg({"cnt" : "sum"})
```

```python
month_dict = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December',
}
bike_df_2011 = bike_df_year.loc["2011"]
bike_df_2011["mnth"] = bike_df_2011.index.map(month_dict)
bike_df_2011
```

```python
bike_df_2012 = bike_df_year.loc["2012"]
bike_df_2012["mnth"] = bike_df_2012.index.map(month_dict)
bike_df_2012
```

```python
# Plot tahun 2011
plt.figure(figsize = (12, 5))
plt.plot(bike_df_2011["mnth"], 
         bike_df_2011["cnt"], 
         marker = "o",
         linewidth = 3
        )
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel("Bulan")
plt.ylabel("Frekuensi Peminjaman")
plt.title("Frekuensi Peminjaman Sepeda per Bulan Tahun 2011", loc = "center", fontsize = 20)
plt.show()

# Plot tahun 2012
plt.figure(figsize = (12, 5))
plt.plot(bike_df_2012["mnth"], 
         bike_df_2012["cnt"], 
         marker = "o",
         linewidth = 3
        )
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel("Bulan")
plt.ylabel("Frekuensi Peminjaman")
plt.title("Frekuensi Peminjaman Sepeda per Bulan Tahun 2012", loc = "center", fontsize = 20)
plt.show()
```

### Penjelasan

* Di tahun 2011, dari bulan Januari terus mengalami peningkatan hingga puncaknya pada bulan Juni, dan stabil hingga bulan Agustus. Mulai di bulan September, peminjaman sepeda mulai mengalami penurunan, hal ini mungkin saja disebabkan oleh faktor-faktor, seperti cuaca, suhu, dan kondisi lainnya.
* Pada tahun 2012, peminjaman sepeda di bulan Februari mengalami sedikit peningkatan dari bulan sebelumnya, dan pada bulan Maret, peminjaman sepeda mengalami kenaikan yang cukup signifikan, yang mungkin saja karena beberapa faktor musim, cuaca, dan lainnya. Kemudian, jumlah peminjaman sepeda terus meningkat secara perlahan dari bulan ke bulan hingga di bulan Oktober, jumlah peminjaman sepeda mengalami penurunan, dan itu berlanjut hingga akhir tahun. Hal ini tentu saja dipengaruhi oleh beberapa faktor, seperti keadaan alam saat itu.

* 2. Bagaimana proporsi peminjaman sepeda berdasarkan cuaca?

```python
bike_df_cuaca = bike_df.groupby(by = "weathersit").agg({
    "cnt" : "sum"
})
bike_df_cuaca
```

```python
plt.figure(figsize = (8,8))
plt.pie(bike_df_cuaca["cnt"], labels = bike_df_cuaca.index, autopct = "%1.1f%%", colors = ['#52acae', '#81acc1', '#afd3d0']) 
plt.axis("equal")
plt.title("Proporsi Peminjaman Sepeda berdasarkan Cuaca")
plt.show()
```

### Penjelasan

* Dari pie chart tersebut, dapat diketahui bahwa presentase atau proposi jumlah peminjaman sepeda paling banyak adalah saat cuaca cerah (clear), yaitu sebanyak 68,6%. Kemudian, disusul oleh cuaca berkabut atau mendung (cloudy) sebanyak 30,3%. Dan terakhir adalah saat cuaca salju ringan atau hujan sebanyak 1,2%.

* 3. Bagaimana proporsi peminjaman sepeda untuk hari kerja dan hari libur?

```python
bike_df_day = bike_df.groupby(by = "workingday").agg({
    "cnt" : "sum"
})
bike_df_day
```

```python
plt.figure(figsize = (8,8))
plt.pie(bike_df_day["cnt"], labels = bike_df_day.index, autopct = "%1.1f%%", colors = ['#145780', '#afd3d0']) 
plt.axis("equal")
plt.title("Proporsi Peminjaman Sepeda berdasarkan Hari Kerja dan Hari Libur", fontsize = 20)
plt.show()
```

### Penjelasan

* Dari pie chart tersebut, dapat diketahui bahwa presentase jumlah peminjaman sepeda saat workingday lebih sedikit daripada saat weekend/holiday, yaitu sebesar 30,4% sementara itu presentase weekend/holiday mencapai 69,6%.

* 4. Bagaimana proporsi peminjaman sepeda berdasarkan musim?

```python
fig, ax = plt.subplots(figsize=(20,10))
colors = ["#90CAF9", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(
    y="cnt",
    x="season",
    data=bike_df.sort_values(by="cnt", ascending=False),
    palette=colors,
    ax=ax
)
ax.set_ylabel("Jumlah Peminjaman", fontsize=33)
ax.set_xlabel("Musim", fontsize = 33)
ax.tick_params(axis="x", labelsize=35)
ax.tick_params(axis="y", labelsize=30)
plt.show
```

* 5. Bagaimana pengaruh suhu dan kecepatan angin terhadap jumlah peminjaman sepeda?

```python
# Menghitung korelasi
correlation_matrix = bike_df[["temp", "windspeed", "cnt"]].corr()

# Membuat Heatmap
plt.figure(figsize = (8,6))
sns.heatmap(correlation_matrix, annot = True, cmap  = "coolwarm", fmt = ".2f")
plt.title("Korelasi antara Suhu, Kecepatan Angin, dan Jumlah Peminjaman Sepeda")
plt.show()
```

### Penjelasan 

* Korelasi atau hubungan antara jumlah peminjaman sepeda ("cnt") dengan suhu ("temp") adalah 0,63. Hal itu menunjukkan bahwa korelasi mereka cukup kuat dan positif. Semakin tinggi suhu, maka jumlah peminjaman sepeda akan cenderung meningkat.
* Korelasi atau hubungan antara jumlah peminjaman sepeda ("cnt") dengan suhu ("windspeed") adalah -0,23. Hal itu menunjukkan bahwa korelasi mereka cukup lemah dan negatif. Semakin tinggi suhu, maka jumlah peminjaman sepeda akan cenderung menurun.
